Заполним пропуски с недостающими данными

```python
print('Медианное значение material: ', data.material.median())
```

```python
data.material = data.material.fillna(1.0, inplace = True)
```

```python
print('Медианное значение state: ', data.state.median())
```

```python
data.state = data.state.fillna(2.0, inplace = True)
```

```python
print('Медианное значение build_year', data.build_year.median())
```

```python
data.build_year = data.build_year.fillna(1979.0, inplace = True)
```

```python
print('Медианное значение floor: ', data.floor.median())
```

```python
data.floor = data.floor.fillna(6.5, inplace = True)
```

```python
print('Медианное значение max_floor: ', data.max_floor.median())
```

```python
data.max_floor = data.max_floor.fillna(12.0, inplace = True)
```

```python
print('Медианное значение num_room: ', data.num_room.median())
```

```python
data.num_room = data.num_room.fillna(2.0, inplace = True)
```
